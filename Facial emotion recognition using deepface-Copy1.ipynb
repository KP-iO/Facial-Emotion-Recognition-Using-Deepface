```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
# read image in 
img = cv2.imread('happyboy.jpeg')
img1 = cv2.imread('b3.jpeg')
```

```python
# TO MAKE COLOUR RGB
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
```

```python
predictions = DeepFace.analyze(img1)
```

```python
# get predictions
predictions = predictions[0]
```

```python
img1 = cv2.imread('happyboy.jpeg')
```

```python
# get the most dominant emotion
predictions['dominant_emotion']
```

```python
# Going to draw rectangle across the face
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

faces = faceCascade.detectMultiScale(gray, 1.1, 4)
for(x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
#                 image.   Drawing BOX          (BGR)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Put text over image
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

# Use PutText method to put text over image or video

cv2.putText(img,
            predictions['dominant_emotion'],
           (0,50),
            font, 3,
            (0,0,255),
            2,
            cv2.LINE_4)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
